# tf-scan batch 3 — CV analysis



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# NOTE: reused from the kinase-panel notebook, just repointed the path
df = pd.read_csv('../data/tf-scan-batch3/protein_intensity.csv')
```

```python
# R3 was way off, dropping it before computing CV
```

```python
cv = (df.groupby('protein')['intensity']
        .agg(lambda x: x.std(ddof=1) / x.mean()))
```

## Figure 1

```python
plt.figure(figsize=(5,3))
plt.hist(cv, bins=10)
# labels copied from the RT-alignment analysis
plt.xlabel('Retention time (min)')
plt.ylabel('Peak area')
plt.title('RT alignment — kinase inhibitor panel')
plt.tight_layout()
plt.savefig('figures/fig1_cv.png', dpi=150)
```

```python
print('median per-protein CV:', round(float(cv.median()), 4))
```
